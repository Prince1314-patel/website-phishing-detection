```python
class_map = {'legitimate':0, 'phishing':1}
df['status'] = df['status'].map(class_map)
```

```python
# correlation matrix

plt.figure(figsize=(15, 10))
correlation_matrix = df.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
status_corr = correlation_matrix['status']
status_corr.shape
```

```python
# finding the columns that correlate to the output column
def feature_selector_correlation(cmatrix, threshold):
    
    selected_features = []
    feature_score = []
    i=0
    for score in cmatrix:
        if abs(score)>threshold:
            selected_features.append(cmatrix.index[i])
            feature_score.append( ['{:3f}'.format(score)])
        i+=1
    result = list(zip(selected_features,feature_score)) 
    return result
```

```python
features_selected = feature_selector_correlation(status_corr, 0.2)
features_selected
```

# Preprocessing

```python
# Setting up predictors and target
selected_features = [i for (i,j) in features_selected if i != 'status']
selected_features
```

```python
X_selected = df[selected_features]
```

```python
X_selected.shape
```

```python
X_selected.head()
```

```python
y = df['status']
y
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    shuffle = True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectFromModel
import optuna
import pickle
import joblib
from sklearn.pipeline import Pipeline
```

```python
def objective(trial):
    # Define the pipeline steps
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('feature_selection', SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42))),
        ('classifier', None)
    ])
    
    # Choose the classifier and its hyperparameters based on trial suggestions
    classifier_name = trial.suggest_categorical('classifier', ['LogisticRegression', 'RandomForest', 'GradientBoosting', 'DecisionTreeClassifier'])
    if classifier_name == 'LogisticRegression':
        classifier = LogisticRegression(
            max_iter=1000,
            C=trial.suggest_float('logistic_C', 1e-5, 1e5, log=True)
        )
    elif classifier_name == 'RandomForest':
        classifier = RandomForestClassifier(
            n_estimators=trial.suggest_int('rf_n_estimators', 50, 200),
            criterion=trial.suggest_categorical('criterion', ['gini', 'entropy','log_loss']),
            max_depth=trial.suggest_int('rf_max_depth', 5, 20)
        )
    elif classifier_name == 'GradientBoosting':
        classifier = GradientBoostingClassifier(
            n_estimators=trial.suggest_int('gb_n_estimators', 50, 200),
            loss=trial.suggest_categorical('loss', ['exponential','log_loss']),
            learning_rate=trial.suggest_float('gb_learning_rate', 1e-3, 1e-1, log=True)
        )
    elif classifier_name == 'DecisionTreeClassifier':
        classifier = DecisionTreeClassifier(
            criterion=trial.suggest_categorical('criterion', ['gini', 'entropy','log_loss']),
            splitter=trial.suggest_categorical('splitter', ['best', 'random']),
            max_depth=trial.suggest_int('rf_max_depth', 5, 20)
        )

    # Set the classifier in the pipeline
    pipeline.set_params(classifier=classifier)
    
    # Fit the pipeline and evaluate performance
    pipeline.fit(X_train, y_train)
    # Perform cross-validation
    
    score = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='accuracy').mean()
    
    return score
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)
```

```python
# Get the best parameters and model
best_params = study.best_params
best_params
```

```python
# Recreate the pipeline with the best parameters
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('feature_selection', SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42))),
    ('classifier', None)
])

classifier_name = best_params['classifier']
if classifier_name == 'LogisticRegression':
    classifier = LogisticRegression(
        max_iter=1000,
        C=best_params['logistic_C']
    )
elif classifier_name == 'RandomForest':
    classifier = RandomForestClassifier(
        n_estimators=best_params['rf_n_estimators'],
        criterion=best_params['criterion'],
        max_depth=best_params['rf_max_depth']
    )
elif classifier_name == 'GradientBoosting':
    classifier = GradientBoostingClassifier(
        n_estimators=best_params['gb_n_estimators'],
        loss=best_params['loss'],
        learning_rate=best_params['gb_learning_rate']
    )
elif classifier_name == 'DecisionTreeClassifier':
    classifier = DecisionTreeClassifier(
        criterion=best_params['criterion'],
        splitter=best_params['splitter'],
        max_depth=best_params['rf_max_depth']
    )
```

```python
# Set the classifier and fit the pipeline
pipeline.set_params(classifier=classifier)
pipeline.fit(X_train, y_train)
```

```python
# Save the fitted pipeline
joblib.dump(pipeline, 'best_model.pkl')
```

```python
from sklearn.metrics import classification_report

# Predict the test set results
y_pred = pipeline.predict(X_test)

# Generate the classification report
report = classification_report(y_test, y_pred, target_names=['legitimate', 'phishing'])
print(report)
```
